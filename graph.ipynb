```python
import pandas as pd
import matplotlib.pyplot as plt

racism = []
sexism = []
xenophobia = []
dates = []
for i in range(7, 9):
    data = pd.read_csv(f'201{i}_predictions.csv')
    racism.append(data['racism'].sum() / len(data))
    sexism.append(data['sexism'].sum() / len(data))
    xenophobia.append(data['xenophobia'].sum() / len(data))
    dates.append("201" + str(i))
    print("Year: 201" + str(i))
    print("Racism:", racism[-1])
    print("Sexism:", sexism[-1])
    print("Xenophobia:", xenophobia[-1])
    print()
for i in range(0, 4):
    data = pd.read_csv(f'202{i}_predictions.csv')
    racism.append(data['racism'].sum() / len(data))
    sexism.append(data['sexism'].sum() / len(data))
    xenophobia.append(data['xenophobia'].sum() / len(data))
    dates.append("202" + str(i))
    print("Year: 202" + str(i))
    print("Racism:", racism[-1])
    print("Sexism:", sexism[-1])
    print("Xenophobia:", xenophobia[-1])
    print()

plt.plot(dates, racism, label='Racism')
plt.plot(dates, sexism, label='Sexism')
plt.plot(dates, xenophobia, label='Xenophobia')

plt.xlabel('Year')
plt.ylabel('Percentage of Tweets')
plt.title('Percentage of Tweets Containing Hate Speech')
plt.legend()

plt.show()
```
